## Account Addresses

```python
acc_adr_url = 'https://api.koios.rest/api/v0/account_addresses'
headers = {'Accept': 'text/csv'}
params= {
    '_stake_addresses': '{stake1uyrx65wjqjgeeksd8hptmcgl5jfyrqkfq0xe8xlp367kphsckq250}',
    '_first_only': 'false',
    '_empty': 'true'
}
```

```python
acc_adr = requests.get(acc_adr_url, headers=headers, params=params)
```

```python
#acc_adr.text.split("\n")[1].split(",")[1:]
```

```python
data = []
for i in range(len(acc_adr.text.split("\n")[1].split(",")[1:])):
    data.append([acc_adr.text.split("\n")[1].split(",")[0], acc_adr.text.split("\n")[1].split(",")[1:][i].replace('\'', '').replace('\"', '').replace('[', '').replace(']', '')])
```

```python
df = pd.DataFrame(data, columns = acc_adr.text.split("\n")[0].split(","))
```

## Address Transactions

```python
#'{' + ', '.join(df['addresses'].to_list()) + '}'
```

```python
data = []
```

```python
for addr in df['addresses'].to_list():
    add_trans_url = 'https://api.koios.rest/api/v0/address_txs'
    headers2 = {'Accept': 'text/csv'}
    params2= {
        '_addresses': f"{'{' + addr + '}'}"
    }
    add_trans = requests.get(add_trans_url, headers=headers2, params=params2)
    for item in add_trans.text.split("\n")[1:]:
        data.append(item.split(","))
```

## Transaction Information

```python
data3 = []
```

```python
for tx in df2['tx_hash'].to_list():
    trans_info_url = 'https://api.koios.rest/api/v0/tx_info'
    params3= {
        '_tx_hashes': f"{'{' + df2['tx_hash'].to_list()[0] + '}'}"
    }
    trans_info = requests.get(trans_info_url, params=params3)
    for output in trans_info.json()[0]['outputs']:
        data3.append([tx, output['payment_addr']['bech32'], output['value']])
```
